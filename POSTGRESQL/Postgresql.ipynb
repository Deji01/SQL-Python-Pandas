# Import Libraries

```python
import psycopg2
import pandas as pd
from sqlalchemy import create_engine
```

## Create Engine

```python
#create sqlalchemy engine
engine = create_engine('postgresql+psycopg2://{user}:{pw}@localhost/{db}'.format(user='postgres',pw='password',db='postgres'))
```

## Read Table

```python
df = pd.read_sql_table('address',engine)
```

```python
df.head()
```

## Read Query

```python
query = '''SELECT * FROM address'''
```

```python
df2 = pd.read_sql_query(query,engine)
```

```python
df2.head()
```

## Read SQL

```python
# Convinience wrapper for read_sql_query & read_sql_table
df3 = pd.read_sql('city',con=engine)
```

```python
df3.tail()
```

```python
query2 = '''SELECT * FROM city'''
```

```python
df4 = pd.read_sql(query2, con=engine)
```

```python
df4.tail()
```

## Joining

```python
query3 = '''SELECT fa.actor_id, fa.film_id, a.first_name, a.last_name, f.title, f.description 
            FROM public.film_actor fa
            JOIN public.actor a
            ON fa.actor_id = a.actor_id
            JOIN public.film f
            ON fa.film_id = f.film_id
            ORDER BY f.title'''
```

```python
df5 = pd.read_sql(query3, engine)
```

```python
df5.head()
```

```python
df5['title'].unique()
```

```python
movie_titles = df5['title'].unique()
```

```python
print('There are unique {} movies in this dataframe'.format(len(movie_titles)))
```

## Save to SQL

```python
df5.to_sql('film_inventory', engine, if_exists='fail')
```
